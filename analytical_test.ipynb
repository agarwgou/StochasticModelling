```python
s_price= np.arange(2,200,2)
delta=BlackScholesDCashCallDelta(s_price, 80, 0, 0.2 ,1) - BlackScholesDCashCallDelta(s_price, 120, 0, 0.2 ,1)

plt.plot(s_price,delta)
plt.ylabel("Delta")
plt.xlabel("Stock Price")
plt.title("Delta of the Digital Payoff")
plt.grid()
plt.show()
```
